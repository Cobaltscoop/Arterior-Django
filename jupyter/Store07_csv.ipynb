# **2019 Image Indexing**
## **1 Store 데이터와 연결된 Index 추가하기**

- 많은 정보의 확인이 필요한 테이블은 남겨두고 나머지도 작업하기

```python
import pandas as pd
table = pd.read_csv('../media/y2019/store.csv', header=None)
table.columns = ['title', 'artist', 'address1', 'address2', 'address3', 'lan', 
    'lon', 'phone', 'work_time', 'before', 'content1', 'content2', 'content3']
print(table.shape)
```

```python
# Artist Code 만들고 추가하기
code_artist = []  
for _ in table['artist']:
    if _ not in code_artist:
        code_artist.append(_)

code_artist = {_:f"A{no+1:02d}"   for no,_ in enumerate(code_artist)}
temp_code = [code_artist[_]  for _ in table['artist']]
table.insert(0, 'code', temp_code)
table.tail(1)
```

```python
code_detail = []
for code in sorted(set(table['code'])):
    data_number = table[table.code==code].shape[0]
    for _ in range(1, data_number + 1):
        code_detail.append(f"{code}{_}")
table.insert(1, 'code_detail', code_detail)
table.to_csv('../media/y2019/store_raw.csv', header=None, index=None)
table.tail(3)
```

```python
list_images = []
for code in table['code_detail']:
    list_temp = []
    for image in ['before', 'after', 'banner', 'img1', 'img2', 'img3',  'process1', 'process2', 'process3', 'map']:
        list_temp.append(f"y2019/store/{code}_{image}.jpg")
    list_images.append(list_temp)
    
len(list_images)
table_image = pd.DataFrame(list_images)
table_image.tail(3)
```

```python
table_merged = pd.concat([table, table_image],axis=1)
table_merged.to_csv('../media/y2019/store_raw_images.csv', header=None, index=None)
table_merged.tail(3)
```
