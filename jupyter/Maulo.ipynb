```python
import requests
url = "http://www.maulo.kr/down/maulo.asp?page=1"
response = requests.get(url)
response = response.content.decode('utf-8')
```

```python
import pandas as pd
table = pd.read_html(response, header=0)[0]
table = table.fillna('')
table.head()
```

## **get Link url**

```python
# table 에 포함된 Link 갯수
from lxml.html import fromstring
response_lxml = fromstring(response)
images = response_lxml.xpath("//a")
images_hrefs = [_.get('href') for _ in images]
images_hrefs = [_.split('/')[-1]   for _ in images_hrefs]
```

```python
import re
i = 0
image_links = []
for length in range(len(table)):
    image_length = table.loc[length, "이미지"]
    image_count = re.findall("그림", image_length)
    if len(image_count) != 0:    
        image_links_temp = []
        for I in range(len(image_count)):
            image_links_temp.append(images_hrefs[i])
            i += 1
        image_links_temp = " // ".join(image_links_temp)
    else:
        image_links_temp = ""
    image_links.append(image_links_temp)
```

```python
table['이미지'] = image_links
table.tail()
```

```python
table.to_excel("page1.xlsx")
```
