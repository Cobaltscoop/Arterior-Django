## **1 Editing the Store Code**
Store Code 정보 추가하기

````python
import pandas as pd
file_name = '../media/csv/backup/store_org.csv'
table_source = pd.read_csv(file_name, header=None)
table_source.tail(3)
````

````python
import re
file_name = '../media/csv/backup/store_code.csv'
token_group, token_mid = re.compile(r"[A-C]{1}"), re.compile(r"-[\d]{1,2}-")
result_code, result_artist = [], []
table_code = pd.read_csv(file_name, header=None)
for _ in table_code[0]:
    group, number_store = "".join(token_group.findall(_)), int(_[-1:])
    number_middle = int("".join(token_mid.findall(_)).replace('-',''))
    code_artist = f"{group}{number_middle:02d}"
    code = f"{code_artist}{number_store}"
    result_artist.append(code_artist); result_code.append(code)
table_code[0] = result_code
# table_code.insert(0,'artist',result_artist)
table_code.head(3)
````

````python
# table_source 에 코드 추가하기
result_artist = []
result_store = []
for _ in table_source[0]:
    instance = table_code[table_code[1] == _]
    index_integer = instance.index.astype(int).values[0]
    # result_artist.append(table_code.loc[index_integer,'artist'])
    result_store.append(table_code.loc[index_integer,0])
table_source.insert(0,'index', result_store)
# table_source.insert(0,'artist', result_artist)
# table_source = table_source.sort_values(['artist','store']).reset_index(drop=True)
table_source = table_source.sort_values('index').reset_index(drop=True)
table_source.head()
````

## **2 Raw Store Code**
Store Code 세부정보 가공하기

````python
import re
import pandas as pd
file_name = '../media/csv/backup/store_text.csv'
tokenizer1 = re.compile(r'[A-C]_[\d]{2}_[\d]{1}')
tokenizer2 = re.compile(r'[A-C]_[\d]{2}')
table_data = pd.read_csv(file_name, header=None)
# Index Name Field 추가하기
table_data.insert(0, 'index', ["".join(tokenizer1.findall(_)).replace('_', '')  for _ in table_data[0]])
# table_data.insert(0, 'artist', ["".join(tokenizer2.findall(_)).replace('_', '')  for _ in table_data[0]])
# 분류 Field 이름으로 변경하기
table_data[0] = [_.replace("".join(tokenizer1.findall(_))+"_", "")  for _ in table_data[0]]
table_data_temp = table_data[table_data['index']=='A113']
table_data_temp
````

````python
resultTable = []
for _ in table_source['index']:
    instance = table_data[table_data['index']== _]
    banner = instance[instance[0] == 'banner_txt_01'][1].values[0].strip()
    title = instance[instance[0] == 'title_01'][1].values[0].strip()
    title_script = instance[instance[0] == 'title_02'][1].values[0].strip()
    image1 = instance[instance[0] == 'image01_txt'][1].values[0].strip()
    image2 = instance[instance[0] == 'image02_txt'][1].values[0].strip()
    image3 = instance[instance[0] == 'image03_txt'][1].values[0].strip()
    resultTable.append([banner, title, title_script, image1, image2, image3])
tableTemp = pd.DataFrame(resultTable)
tableTemp.columns = ["banner", "title", "script", "text1", "text2", "text3"]
table_store = pd.merge(table_source, tableTemp, left_index=True, right_index=True,how='right').iloc[:,1:]
# table_store.to_csv("../media/csv/store.csv", header=None, index=None)
table_store.tail()
````
