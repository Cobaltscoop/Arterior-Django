- Store DataTable 에 추가될 **Image 내용** 확인하기 

## **1 Reading the Store Code**
앞에서 작업 완료한 Table 불러오기

```python
import pandas as pd
file_name = '../media/csv/store_raw.csv'
table_csv = pd.read_csv(file_name, header=None)
table_csv.head(3)
```

## **2 Checking & Building The file List**
DB 에 필요한 파일들 목록을 생성하고 내용 확인하기

```python
import os
os.path.isfile('../media/store/A011_banner1.jpg')
```

```python
from glob import glob
text_file = '*.jpg'
text_path = '../media/store/'
file_list = sorted(glob(text_path + text_file))

# Code 를 호출하며 파일 이름을 만들고, 존재여부 확인하기
result_list = []
text_path_django = 'store/'
list_check = ['_banner1.jpg','_banner2.jpg','_image_101.jpg','_image_102.jpg','_image_103.jpg',
    '_image_201.jpg','_image_202.jpg','_image_203.jpg','_image_301.jpg','_image_302.jpg','_image_303.jpg']

for _ in table_csv[0]:
    list_temp = []
    for check in list_check:
        file_name = text_path + _ + check
        if os.path.isfile(file_name) == False:
            print(f"{file_name} is not existed"); break
        else:
            file_name = text_path_django + _ + check
            list_temp.append(file_name)
    result_list.append(list_temp)
tableTemp = pd.DataFrame(result_list, columns=list_check)
tableTemp.head()
```

## **3 Merging The Table**
새로 추가한 Table 을 연결하기

```python
table_store = pd.merge(table_csv, tableTemp, left_index=True, right_index=True,how='right').iloc[:,1:]
# table_store.to_csv('../media/csv/store.csv', index=None, header=None)
print(table_store.head(3))
```
